```python
# index for selecting data 0.75 is the percentage in training
index=np.array([1 if np.random.random() < 0.75 else 0 for i in range(len(df))])
X_train=np.array(df[df.columns[0]])[index==1]
X_test=np.array(df[df.columns[0]])[index==0]
Y_train=np.array(df[df.columns[1]])[index==1]
Y_test=np.array(df[df.columns[1]])[index==0]
```

```python
#Normalizamos los datos
mediaXtr=np.mean(X_train)
sdXtr=np.std(X_train)
mediaYtr=np.mean(Y_train)
sdYtr=np.std(Y_train)
X_train = (X_train-mediaXtr) / sdXtr
X_test = (X_test-mediaXtr) / sdXtr
Y_train = (Y_train-mediaYtr) / sdYtr
Y_test = (Y_test-mediaYtr) / sdYtr
```

```python
plt.scatter(X_train,Y_train)
```

```python
def RegLinLMS(X,Y,W,eta):
    w_act = W
    for j in range(100):
        for i in range(len(X)):
            #import pdb; pdb.set_trace()
            Vxi = np.dot(w_act , [1, X[i]])
            error = Y[i] - Vxi
            w_act[0] = w_act[0] + eta * ( error ) 
            w_act[1] = w_act[1] + eta * ( error ) * X[i]
            
            
    return w_act    
```

```python
W = RegLinLMS(X_train,Y_train,[1,1],0.05)
```

```python
x = np.linspace(X_test.min(),X_test.max(),100)
y = x*W[1]+W[0]
```
